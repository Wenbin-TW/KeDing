```python
import pyodbc
import tkinter as tk

def submit_cookies():
    global cookies_string
    cookies_string = text_input.get("1.0", tk.END).strip()  
    root.destroy()  

server = ''
database = 'bidb'
username = 'kdmis'
password = ''

connection_string = (
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    f'UID={username};'
    f'PWD={password}'
)

root = tk.Tk()
root.title("Cookies 输入")
root.geometry("600x500")  
label = tk.Label(root, text="请输入 Cookies：", font=("Arial", 14))
label.pack(pady=10)
text_input = tk.Text(root, wrap=tk.WORD, font=("Arial", 12), height=20, width=70)
text_input.pack(padx=10, pady=10)
submit_button = tk.Button(root, text="提交", command=submit_cookies, font=("Arial", 12))
submit_button.pack(pady=10)

root.bind('<Return>', lambda event: submit_cookies())
root.mainloop()
if not globals().get("cookies_string"):
    print("未输入 Cookies，程序退出。")
    exit()

try:
    conn = pyodbc.connect(connection_string)
    cursor = conn.cursor()
    insert_query = """
        INSERT INTO dbo.HR_COOKIES (indeed, black_1, black_2, black_3)
        VALUES (?, NULL, NULL, NULL)
    """
    cursor.setinputsizes([(pyodbc.SQL_WVARCHAR, len(cookies_string))])  # 明確設置類型
    cursor.execute(insert_query, cookies_string)
    conn.commit()
    print("Cookies 插入成功！")

except Exception as e:
    print(f"發生錯誤：{e}")

finally:
    if 'cursor' in locals() and cursor:
        cursor.close()
    if 'conn' in locals() and conn:
        conn.close()
```
